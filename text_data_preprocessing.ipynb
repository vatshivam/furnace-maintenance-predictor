```python
import numpy as np
import pandas as pd
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import random 
from nltk.corpus import stopwords
import string
import nltk
import re
```

```python
df = pd.read_csv('headlines.csv',index_col=0)
```

```python
df.headlines = df.headlines.str.replace("com","").str.replace("’","").str.replace("‘","")
```

```python
df.headlines.to_list()
```

```python
df.headlines = df.headlines.str.lower()
```

```python
df.headlines = df.headlines.str.split()
df.headlines
```

```python
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
def get_simple_pos(pos):
    if pos.startswith('J'):
        return wordnet.ADJ
    elif pos.startswith('N'):
        return wordnet.NOUN
    elif pos.startswith('V'):
        return wordnet.VERB
    elif pos.startswith('R'):
        return wordnet.ADV
    else: return wordnet.NOUN
```

```python
sp_char='[@_!#$%^&*()<>?/\|}{~£+:.-]™"'
lemmatizer = WordNetLemmatizer()
def clean_data(sentences):
    clean_sentences=[]
    for sentence in sentences:
        clean_sentence = []
        for word in sentence:
            w = word.strip()
            if (w not in stops) and (w not in sp_char) and w:
                pos = pos_tag([w])[0][1]
                clean_word = lemmatizer.lemmatize(w,pos=get_simple_pos(pos))
                clean_sentence.append(clean_word)
        clean_sentences.append(clean_sentence)
    return clean_sentences
```

```python
clean_data = clean_data(df.headlines.to_list())
```

```python
transactions=[]
for string in clean_data:
    transactions.append(','.join(string))
```

```python
pd.DataFrame({'transactions':transactions}).to_csv("transactions.csv")
```
