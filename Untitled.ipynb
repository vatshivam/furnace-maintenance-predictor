```python
data = pd.read_csv("./data/classification_1.csv")
```

```python
df = data[data.columns.values[:6]]
```

```python
df = df.set_index("timestamp")
```

```python
import statsmodels.api as sm
```

```python
y = df['Good/Bad']
x = df.drop(columns=['Good/Bad'])
```

```python
x.values
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
for i in range(4):
    print("P-value",adfuller(x.values[:,i])[1])
```

From the above output, we can infer that the predictors are stationary.

```python
for i in range(4):
    sm.graphics.tsa.plot_acf(x.values[:,i], lags=300)
```

```python
for i in range(4):
    sm.graphics.tsa.plot_pacf(x.values[:,i], lags=30)
```

From the above ACF plots, we can see that the distance between consecutive crests and troughs is around 200 units. This is true for first, third and fourth feature. The second feature does not display autocorrelation. Hence including 200 lagged features is a possible solution for modelling. However, this will increase the feature space tremendously.

Trying differencing to check if it reduces the number of significant lags

```python
# df['B_18_diff'] = df['B_18'].diff(200)
# df['B_19_diff'] = df['B_19'].diff(200)
# df['B_16_diff'] = df['B_16'].diff(200)
```

```python
for i in range(3):
    sm.graphics.tsa.plot_acf(df[['B_18_diff','B_19_diff','B_16_diff']].values[200:,i], lags=1000)
```

Proceeding with lags/4 number of lags for modelling the 3 features B_19, B_18 and B_16

```python
for i in range(1,200,4):
    x['B_16_'+str(i)] = x['B_16'].shift(i)
    x['B_19_'+str(i)] = x['B_19'].shift(i)
    x['B_18_'+str(i)] = x['B_18'].shift(i)
```

```python
df = pd.concat([x.reset_index(),data[data.columns.values[5:]]],axis=1)
df = df.set_index('timestamp',drop=True)
df.head()
```
